###### 4. Logical vs. Physical Diagrams
***

In this presentation we'll take a look at two types of diagrams often used in network planning and documentation. The physical and the logical. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Physical Diagrams
***

The physical is concerned with the actual physical arrangement of all of the components that make up a network.

So all cables, hardware, routers, switches, servers, even the desktops and the printers. 

In essence, every hardware device that will participate in the network should be included in a physical diagram. 

Now this might end up resembling a wiring schematic or possibly even a blueprint, but the idea is that the emphasis is placed on the endpoints. 

> Where are things actually located? 

And then you can certainly indicate the cable runs between them, but it does represent the way that everything will physically be structured. 

So this is useful, of course, when planning a network from scratch or when planning to replace outdated equipment or any equipment that might actually be hampering the performance for any given reason. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Logical Diagrams
***

The logical is not really so much concerned with any specific device or where it's placed, but rather it describes the manner in which information flows through a network.

So for example, which devices are able to communicate with one another, and by which means. 

So you might still see some indications of certain types of devices but it's more so describing the flow of traffic, so, in this example, we see the Internet separated by a router, then we see a bit of a demilitarized zone where there are three different subnets, then a firewall separating them from our internal networks. 

But, the idea is that this is simply laying out the flow of information, we don't see specific servers or workstations, or cabling or anything along those lines. 

So typically, what you'll see might be the subnets, VLAN IDs, the subnet masks and the addresses that are used within those subnets, and of course the size of any given subnet and, as mentioned, you might still see certain types of networking devices, including firewalls, routers and switches, but that's dependent on the diagram.

But, you know, they aren't really used to indicate that this is where the physical device is located, but rather this is simply its purpose, its function. 

So you might also see some of the routing protocols indicated in a logical diagram, so that you know that router A is communicating with router B using OSPF or whichever routing protocol you choose to implement. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Construct both Physical and Logical
***

So again, the idea is to really construct both so that you do understand how everything is laid out, where the devices are placed, but then you also want to indicate the overall flow of information, so the two of them together give you a much better picture of everything that will ultimately be implemented.

###### 5. Rack Diagrams
***

In this presentation, we'll talk about the simple computer rack, or more commonly referred to as just a rack. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Rack
***

The reason why is because in larger more structured environments, most of the hardware tends to be centralized in a rack. 

And as such, a lot of the diagramming actually focuses on how the racks are constructed. 

Now the rack itself is little more than what it sounds like, it's a metal frame used to hold various hardware devices, including servers, hard disks, routers, and switches. 

Now, if you are fairly new to IT and just getting started with networking, you might think how does a server fit into a rack, it's just a computer like any other one, just a little more robust and that is true, but in many cases, in the larger environments, the servers are little more than the processors, the memory, and maybe the network adapters.

Then just about everything else is connected remotely, so the storage is accessed remotely, you administer it remotely by using things like Remote Desktop, so as such, those types of servers can absolutely just be rack mounted. 

They're often referred to as blade servers, and again, they can sit right in there along with racks of hard disks, and racks of routers, and switches, so they can all be centralized in these racks. 

Now, with respect to actually mounting the devices, typically, they are screwed into the front. 

They usually have tabs that have screw holes in them that line up with screw holes on the rack and you, quite simply, screw the device in, and then all of the wiring is done often on the back side for servers, but for things like switches, it would be on the front side. 

But a standard traditional rack is 19 inches wide, and typically built using aluminium or stainless steel but you know that depends on how robust you need it to be, with respect to the amount of weight it can hold. 

Aluminium, of course, is much lighter, it would not be as strong, so if you needed to have a lot of devices in there, you might want to move up to stainless steel. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Rack Diagrams
***

But, as mentioned, [Rack Diagrams] this is a key component of diagramming because a lot of racks have servers, routers, switches, various other types of devices in them. 

The rack diagram allows you to visualize how the devices will be mounted, so, again, with respect to the servers, the computers that are in there, and the networking equipment, you want to be able to label all of this and, in essence, map out how they are all going to be configured. 

So you still need to document the networking devices so that you can quickly determine which equipment is what when you are purchasing and, of course, where they need to go in the rack. 

So if, for example, you have enough storage to fill up an entire rack then you kind of have an idea of how many devices you need to purchase and then if you need x number of blade servers and x number of switches, you know what to purchase and you know how much space you have within any given rack.

So, again, it just makes things a lot easier to visualize when designing everything and because of the fact that everything is centralized in the rack, the rack diagram is a significantly important component in terms of documenting how everything is assembled.

###### 6. Change Management Documentation
***

Of course well all know that nothing stays the same and when it comes to implementing changes in any given IT environment, there is often what's known as a change management process. 

So in this presentation, we will take a look at how you can implement change process management and take a look at the associated documentation as well.

###### Change Management Process
***

So typically, when it comes to any kind of change, there is usually a formal process which manages modifications to various programs and systems. 

You generally just don't stroll in and start making changes without some kind of approval, and of course, some kind of official process. 

Now, the process is typically handled by change review boards, again, usually in the more structured environments. 

You might not see this all that much in a small business but stakeholders are typically able to voice any concerns prior to the changes actually being made and details with respect to the changes should be thoroughly documented. 

So that the stakeholders do understand what it is that is going to be changed and of course, what the impact of those changes will be. 

So it's also important to have a roll back plan because some changes can have unexpected and negative results. 

So again, you generally don't just want to jump into a change without having some kind of a plan to say well, we can go back to the way things were in the event that this change does not prove to be positive.

So formal documentation is typically vital and that can come in a variety of forms, including secured email, paper memos and in the more structured environments, you might even see what's known as the Request for Change document, or RFC.

I can go into the wiring closet and see, well, there are all of my data ports but where are the other ends of them? because if you are in an environment where every single connection terminates within the wiring closet but then you have, let's say, 50 offices. If it's an entire switch that maybe went down, you can quickly and easily move those connections to a different switch, for example. 

So you know again, it's really just all about knowing where things are. 

It just solves a lot of problems with respect to not having to trace this down to the other end because if you don't know where the other end of this connects, clearly you are going to waste a lot of time just trying to find that out, then you can start your troubleshooting. 

But effective diagramming of the port locations and all of the wiring and where the cables are run will certainly make things easier when it comes to trying to run down any type of communication

***
###### 8. Intermediate Distribution Frames (IDF) and Main Distribution Frame (MDF) documentation
***

In this presentation, we'll take a look at two wiring solutions known as the MDF and the IDF. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Main Distribution Frame (MDF)
***

Start with the M, which is the Main Distribution Frame.

Quite simply, this interconnects and manages telecommunication wiring between the MDF itself and then any number of the IDFs, which are the intermediate distribution frames. 

Essentially, this is a big split point. 

The MDF is typically where service comes into a building and then you split it from that point into as many as are necessary IDFs to take that service and carry it to different floors of the building or even two different buildings, if, say, for example, it's on a campus or something along those lines. 

So essentially, it just allows you to connect the lines that are coming into the building with your own internal network, so it really is just the meeting point and, as sort of mentioned, the split point between the public and the private. 

So if we are talking about just Internet service, for example, the public would be the Internet service provider, providing you with the connection but then, from that point, it has to split off to all of the other locations that need to be serviced, and that's where the IDFs come in. 

So again, this is typically a larger scale implementation than, let's say, something like patch panels whereas you tend to see patch panels just maybe splitting off cables to several offices, maybe on the same floor. 

This is the same kind of idea, but really just on a larger scale, so it can provide for greater flexibility for telecommunication facilities. 

It, again, can be used to split off to multiple floors or even to separate buildings. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Main Distribution Frame (MDF) - Characteristics
***

So in terms of the characteristics of an MDF itself, it really is just a little more than a long steel rack that is accessible from both sides. 

So workers can get at both the front and the back and then it offers horizontal termination blocks that are typically accessed on the front of the rack shelves and then jumpers that are typically used to interconnect, again, that main public in to the split points out and depending on the size, they can hold hundreds of thousands of jumpers to establish as many connections as are necessary. 

But again, of course, that depends on the scale. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Intermediate Distribution Frames (IDF)
***

But the IDFs, again, come in to play when you simply need to create these distribution points. 

So it's a bit of a hierarchy, if you will. You see all of the individual workstations connected to the IDFs then there are several IDFs connected to the MDF. 

So, again, it's not unlike a switch, per se but we are just talking about getting the cabling to those systems itself, so it is still a cable rack, and it manages the telecommunications wires between the MDF and the workstation devices.

So again, you kind of see this hierarchy where the MDF would kind of represent a single connection point, then you might have, let's say, three or five or ten IDFs and then, of course, you can connect all of the individual work stations to the IDFs in those buildings or on those floors, or however you feel it is necessary to implement. 

But it's just working your way from that main primary connection point down to the individual workstation. 

So they connect, typically, your internal lines to the MDF. 

So again looking at an example, an organization may comprise of multiple buildings on a single compound, such as a university. 

They may also house one MDF in the main building and then IDFs in each of the other buildings because you would not need more than one MDF. 

That services everything then you split off to the other buildings, and you have possibly even just a single IDF in any given building, depending on its capacity.

But the idea is to simply create this hierarchical distribution so that we don't have multiple MDFs all over the place. 

Just one of those, then a few more IDFs, then they branch off to the individual systems themselves.

###### Baseline - Normal

You can try to correlate that with some kind of event that occurred, such as, when we implemented this new application or this new device, we saw something outside of the ordinary.
